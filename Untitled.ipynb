```python
def detect_outliers_iqr(data):
    
    numeric_data = data.select_dtypes(include=['number'])
    outlier_counts = {}

    for col in numeric_data.columns:
        Q1 = numeric_data[col].quantile(0.25)
        Q3 = numeric_data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outlier_count = ((numeric_data[col] < lower_bound) | (numeric_data[col] > upper_bound)).sum()
        outlier_counts[col] = outlier_count

    return pd.DataFrame.from_dict(outlier_counts, orient='index', columns=['Количество выбросов'])

def stat_info(data):
    display(data.sample(7))
    print('_' * 80, '\n')
    display(data.info())
    print('_' * 80, '\n')
    display(f'Количество дубликатов: {data.duplicated().sum()}')
    print('_' * 80, '\n')
    display(data.describe())

    # Проверка на выбросы
    outliers = detect_outliers_iqr(data)
    print('_' * 80, '\n')
    print("Выбросы в данных:")
    display(outliers)

    # Гистограммы
    data.hist(bins='auto', figsize=(12, 8), color='black', edgecolor='white', grid=True)
    plt.suptitle("Распределение числовых данных", fontsize=14, fontweight='bold')
    plt.show()

    # Ящики с усами (boxplot)
    numeric_data = data.select_dtypes(include=['number'])
    fig, axes = plt.subplots(nrows=len(numeric_data.columns), ncols=1, figsize=(12, 4 * len(numeric_data.columns)))

    for ax, col in zip(axes, numeric_data.columns):
        sns.boxplot(x=numeric_data[col], ax=ax, color='gray')
        ax.set_title(f"Ящик с усами для {col}", fontsize=12, fontweight='bold')

    plt.tight_layout()
    plt.show()

def corr_matrix(data):    
    numeric_data = data.select_dtypes(include=['number'])  # Оставляем только числовые столбцы
    corr_matrix = numeric_data.corr()
    display(corr_matrix)  # Выводим числовую матрицу корреляций
    fig, ax = plt.subplots(figsize=(15, 6))
    sns.heatmap(corr_matrix, annot=True, cmap="Greys", fmt=".2f", linewidths=0.5, ax=ax)
    plt.show()
```

### Обучение и проверка модели

```python
dfs = {'df_0': df_0, 'df_1': df_1, 'df_2': df_2}
results = []

for name, df in dfs.items():
    print(f"Обработка: {name}")
    
    features = df.drop(columns=['product', 'id']) 
    target = df['product']

    features_train, features_valid, target_train, target_valid = train_test_split(features, target, test_size=0.25, random_state=42)

    # Масштабируем данные
    scaler = StandardScaler()
    features_train = scaler.fit_transform(features_train)
    features_valid = scaler.transform(features_valid)

    # Обучаем модель
    model = LinearRegression()
    model.fit(features_train, target_train)
    
    # Предсказания
    predicted_valid = pd.Series(model.predict(features_valid), index=target_valid.index)
    
    # Расчет метрик
    mse = mean_squared_error(target_valid, predicted_valid)
    rmse = np.sqrt(mse)

    mean_pred = predicted_valid.mean()

    print(f"Средний запас предсказанного сырья: {mean_pred:.2f}")
    print(f"RMSE модели: {rmse:.2f}\n")

    # Сохраняем результаты
    results.append({'dataset': name, 'mean_pred': mean_pred, 'rmse': rmse})

# Выводим общий анализ результатов
results_df = pd.DataFrame(results)
results_df
```

### Подготовка к расчёту прибыли

```python
BUDGET = 10**10
COST = 450000
AMOUNT = 200

state = np.random.RandomState(42)

def bootstrap(predicted, fact, region_name):
    values = []
    for i in range(1000):
        # Выбираем 500 случайных скважин и выбираем 200 лучших по предсказаниям
        top_200_predict = predicted.sample(n=500, random_state=state).sort_values(ascending=False).head(200)
        top_200_fact = fact[top_200_predict.index]
        # Расчёт прибыли
        profit = top_200_fact.sum() * COST - BUDGET
        values.append(profit)
    values = pd.Series(values)
    
    # Расчёт доверительного интервала и риска убытков
    lower = values.quantile(q=0.025)
    upper = values.quantile(q=0.975)
    region_mean = values.mean()
    risk = (values < 0).mean()
    
    # Вывод результатов
    print('Средняя прибыль региона составляет {:.1f} руб.'.format(region_mean))
    print('95% доверительный интервал: {:.1f} - {:.1f} руб.'.format(lower, upper))
    print('Риск убытков - {:.1%}'.format(risk))
    
    # Построение гистограммы
    plt.figure(figsize=(10, 6))
    plt.hist(values, bins='auto', color='black', edgecolor='black')
    plt.axvline(lower, color='red', linestyle='--', label='Нижняя граница 95% интервала')
    plt.axvline(upper, color='green', linestyle='--', label='Верхняя граница 95% интервала')
    plt.axvline(region_mean, color='blue', linestyle='-', label='Средняя прибыль')
    plt.title(f'Распределение прибыли для региона {region_name}')
    plt.xlabel('Прибыль, руб.')
    plt.ylabel('Частота')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    return values

# Для каждого региона вызываем bootstrap
for name, df in dfs.items():
    print(f"\nОбработка для региона: {name}")
    
    features = df.drop(columns=['product', 'id']) 
    target = df['product']

    # Разделение данных на обучающую и валидационную выборки
    features_train, features_valid, target_train, target_valid = train_test_split(
        features, target, test_size=0.25, random_state=42
    )

    # Масштабируем данные
    scaler = StandardScaler()
    features_train = scaler.fit_transform(features_train)
    features_valid = scaler.transform(features_valid)

    # Обучаем модель
    model = LinearRegression()
    model.fit(features_train, target_train)
    
    # Предсказания для валидационной выборки
    predicted_valid = pd.Series(model.predict(features_valid), index=target_valid.index)
    
    # Применяем bootstrap для валидационной выборки
    values = bootstrap(predicted_valid, target_valid, name)
```
